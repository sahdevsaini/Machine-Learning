## ---------------- Ploting Best fit line in Linear Regression -----------------

#### Importing Libraries

````python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
````

#### Generate 'random' data

````python
np.random.seed(0)
X = 2.5 * np.random.randn(100) + 1.5   # Array of 100 values with mean = 1.5, stddev = 2.5
res = 0.5 * np.random.randn(100)       # Generate 100 residual terms
y = 2 + 0.3 * X + res                  # Actual values of Y
````

#### Create pandas dataframe to store our X and y values

````python
df = pd.DataFrame(
    {'X': X,
     'y': y}
)
````

#### Show the first five rows of our dataframe

````python
df.head()
````

#### Equation of best fit line
```python
y = mx + b is the slope intercept form of writing the equation of a straight line. In the equation 'y = mx + b', 'b' is the point, where the line intersects the 'y axis' and 'm' denotes the slope of the line. The slope or gradient of a line describes how steep a line is. It can have either a positive or a negative value.
```


#### Calculate the mean of X and y

````python
xmean = np.mean(X)
ymean = np.mean(y)
````

#### Calculate the terms needed for the numator and denominator of slope(m)


````python
df['xycov'] = (df['X'] - xmean) * (df['y'] - ymean)
df['xvar'] = (df['X'] - xmean)**2
````

#### Calculating Slope of Equation(m)

````python
m = df['xycov'].sum() / df['xvar'].sum()
````

#### Intercept of Equation(b)



````python
b = ymean - (m * xmean)
````

````python
print(f'intercept(b) = {b}')
print(f'Slope(m) = {m}')
````

````python
ypred =  m*X  + b
````

````python
# Plot regression against actual data
plt.figure(figsize=(12, 6))
plt.plot(X, ypred)     # regression line
plt.plot(X, y, 'ro')   # scatter plot showing actual data
plt.title('Actual vs Predicted')
plt.xlabel('X')
plt.ylabel('y')

plt.show()
````
